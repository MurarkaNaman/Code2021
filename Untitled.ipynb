```python
theBoard = {'1' : ' ' ,'2' : ' ' ,'3' : ' ' ,
            '4' : ' ' ,'5' : ' ' ,'6' : ' ' ,
            '7' : ' ' ,'8' : ' ' ,'9' : ' ' ,}

win = [['1' , '2' , '3'] , ['4' , '5' , '6'] , ['7' , '8' , '9'] ,
       ['1' , '5' , '9'] , ['3' , '5' , '7'] , ['1' , '4' , '7'] , 
       ['2' , '5' , '8'] , ['3' , '6' , '9'] ]


def printBoard(board) : 
    print(board['1'] + '|' + board['2'] + '|' + board['3'])
    print('-+-+-')
    print(board['4'] + '|' + board['5'] + '|' + board['6'])
    print('-+-+-')
    print(board['7'] + '|' + board['8'] + '|' + board['9'])

def game() : 
    
    turn = 'X'
    count = 0 
    for i in range(10) :
        
        printBoard(theBoard)
        print("It's your turn " , turn)
        
        move = input()
        
        try : 
            if theBoard[move] == ' ' :
                theBoard[move] = turn
                count += 1 
            else : 
                print("Sorry the move is already there")
                continue
        except : 
            print("you are fucking asked to enter a number")
            continue
            
        if count >= 5 :
            for i in range(len(win)):
                if theBoard[win[i][0]] == theBoard[win[i][1]] == theBoard[win[i][2]] != ' ':
                    printBoard(theBoard)
                    print('********************')
                    print(turn , " Wins the Game")
                    print('********************')
                    return  
                else : 
                    continue

        if turn == 'X' :
            turn = 'O'
        else : 
            turn = 'X'
        
        if count == 9 : 
            print("Sorry it's a tie")
            
game()
```
